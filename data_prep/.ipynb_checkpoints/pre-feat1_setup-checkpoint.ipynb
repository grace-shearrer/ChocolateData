```python
## Derivatives setup 
import glob, os 
from shutil import copy2
```

```python


basedir = '/projects/niblab/bids_projects/Experiments/ChocoData'
fprepdir = os.path.join(basedir, "fmriprep")
derivdir= os.path.join(basedir, "derivatives")
SUBJECTS = []
for sub_dir in sorted(glob.glob(os.path.join(fprepdir, "sub-*"))):
    sub_id = sub_dir.split("/")[-1]
    SUBJECTS.append(sub_id)
SESS = "ses-3"


# get anat 
#/projects/niblab/bids_projects/Experiments/ChocoData/fmriprep/sub-001/ses-2/fmriprep/sub-001/anat
for sub_id in SUBJECTS:
    anat_file = os.path.join(fprepdir, sub_id, SESS, "fmriprep/%s/anat/%s_T1w_space-MNI152NLin2009cAsym_preproc.nii.gz"%(sub_id, sub_id))
    #print(anat_file)
    deriv_path = os.path.join(derivdir, sub_id, SESS, "anat")
    
    try:
        copy2(anat_file, deriv_path)
        print("COPYING FILE %s >>>-------> %s"%(anat_file, deriv_path))
    except:
        print("******%s has no files"%sub_id)
for sub_id in SUBJECTS:
    org_file = os.path.join(derivdir, sub_id, SESS, "anat/%s_T1w_space-MNI152NLin2009cAsym_preproc.nii.gz"%(sub_id))
    new_file = os.path.join(derivdir, sub_id, SESS, "anat/highres.nii.gz")
    try:
        os.rename(org_file, new_file )
    except:
        pass
    
for sub_id in SUBJECTS:
    new_dir = os.path.join(derivdir, sub_id, SESS, "func/Analysis/feat1")
    os.makedirs(new_dir)
    
    
    
milk_evs_dir = "/projects/niblab/data/eric_data/ev_files/milkshake"
imagine_evs_dir = "/projects/niblab/data/eric_data/ev_files/imagine"

milkA_evs = glob.glob(os.path.join(milk_evs_dir, "mkA*.ev"))
milkB_evs = glob.glob(os.path.join(milk_evs_dir, "mkB*.ev"))
milkC_evs = glob.glob(os.path.join(milk_evs_dir, "mkC*.ev"))
milkD_evs = glob.glob(os.path.join(milk_evs_dir, "mkD*.ev"))

# get onsets--milkshake
for sub_id in SUBJECTS:
    milk_ids = []
    #get milkshake IDs 
    #/projects/niblab/bids_projects/Experiments/ChocoData/derivatives/sub-001/ses-2/func
    milkshake_files=glob.glob(os.path.join(derivdir, sub_id, SESS, "func/*milkshake*"))
    #print(milkshake_files)
    for file in milkshake_files:
        for word in file.split("_"):
            if "milkshake" in word:
                milk_id = word.split("-")[1].split("e")[1]
                #print(milk_id)
                if milk_id not in milk_ids:
                    milk_ids.append(milk_id)
    print(sub_id, milk_ids)
    dest=os.path.join(derivdir, sub_id, SESS, "func/onsets")
    for mlk_id in milk_ids:
        if mlk_id == "A":
            for file in milkA_evs:
                copy2(file, dest)
                print("copying file %s"%(dest))
        elif mlk_id == "B":
            for file in milkB_evs:
                copy2(file, dest)
                print("copying file %s"%(dest))
        elif mlk_id == "C":
            for file in milkC_evs:
                copy2(file, dest)
                print("copying file %s"%(dest))
        elif mlk_id == "D":
            for file in milkD_evs:
                copy2(file, dest)
                print("copying file to %s "%(dest))
        else:
            pass
                

            
# rename anat 
```
